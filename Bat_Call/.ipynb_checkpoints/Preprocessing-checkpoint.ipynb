```python
import librosa, librosa.display
import matplotlib.pyplot as plt 
import numpy as np

# To play audio sounds
from pydub import AudioSegment
from pydub.playback import play
from pydub.effects import low_pass_filter, high_pass_filter
```

# Listen to audio file with bat call before and after

```python
### FUNCTIONS ### 

# credit: https://stackoverflow.com/questions/51434897/how-to-change-audio-playback-speed-using-pydub
def speed_change(sound, speed=1.0):
    # Manually override the frame_rate. This tells the computer how many
    # samples to play per second
    sound_with_altered_frame_rate = sound._spawn(sound.raw_data, overrides={
        "frame_rate": int(sound.frame_rate * speed)
    })

    # convert the sound with altered frame rate to a standard frame rate
    # so that regular playback programs will work right. They often only
    # know how to play audio at standard frame rate (like 44.1k)
    return sound_with_altered_frame_rate.set_frame_rate(sound.frame_rate)
```

```python
sound = AudioSegment.from_wav('./Data/20190617_080000.WAV')
five_seconds = sound[:5000]
play(five_seconds)

### BAND PASS ### 
signal_1 = low_pass_filter(sound, 210000)
signal_2 = high_pass_filter(signal_1, 5000)

slow_sound = speed_change(signal_2, 0.1)
play(slow_sound[:5000])
```

```python
### FFT ### 
fft = np.fft.fft(signal)

magnitude = np.abs(fft) # magnitude of each frequency 
frequency = np.linspace(0, sr, len(magnitude)) 

left_frequency = frequency[:int(len(frequency)/2)]
left_magnitude = magnitude[:int(len(frequency)/2)]

echoloc_frequency = frequency[5000:210000]
# Bat echolocation calls vary between 5kHz to 210kHz according to 
# Vassilios Stathopoulos (2018): Bat echolocation call identification 
# for biodiversity monitoring. 
echoloc_magnitude = magnitude[5000:210000]

# Plotting Power Spectrum 

plt.plot(frequency, magnitude)
plt.xlabel('Frequency') 
plt.ylabel('Magnitude')
plt.show() 
```

```python
plt.plot(left_frequency, left_magnitude)
plt.xlabel('Frequency') 
plt.ylabel('Magnitude')
plt.show()
```

```python
plt.plot(echoloc_frequency, echoloc_magnitude)
plt.xlabel('Frequency') 
plt.ylabel('Magnitude')
plt.show()
```

```python
### STFT to Spectrogram ### 
n_fft = 2048 
# number of samples (window we are considering when completing one FFT)
hop_length = 512 

stft = librosa.core.stft(signal, hop_length = hop_length, n_fft = n_fft)
spectrogram = np.abs(stft)

log_spectrogram = librosa.amplitude_to_db(spectrogram)

# librosa.display.specshow(spectrogram, sr=sr, hop_length=hop_length) # visualize spectrogram like data

librosa.display.specshow(log_spectrogram, sr=sr, hop_length=hop_length) # visualize spectrogram like data=
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar()
plt.show()
```

```python
### MFCC ###
MFCCs = librosa.feature.mfcc(signal, n_fft=n_fft, hop_length = hop_length, n_mfcc=13)

librosa.display.specshow(MFCCs, sr=sr, hop_length=hop_length) # visualize spectrogram like data=
plt.xlabel("Time")
plt.ylabel("MFCC")
plt.colorbar()
plt.show()
```
